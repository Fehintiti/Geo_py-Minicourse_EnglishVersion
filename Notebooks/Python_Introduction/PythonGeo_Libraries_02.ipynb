## Matplotlib  Seaborn

Matplotlib is a library created to make static, animated, interactive visualizations with Python. [The User Guide can be found here.](https://matplotlib.org/stable/users/index).

The library [Seaborn](https://seaborn.pydata.org/),on the other hand, is another library based on Matplotlib. It provides a high-level interface for creating informative and statistical graphics.

```python
# Let's import a specific module from the Matplotlib library
import matplotlib.pyplot as plt

# Importing seaborn
import seaborn as sns

# Importing numpy to create the data
import numpy as np

# Importing os to deal with folders when saving
import os

# Magic function to display plots in the notebook
%matplotlib inline
```

For the visualization examples, we will use the main functions:

- `subplots`: creates space for one or more figures in rows and/or columns.
- `scatter`: Creates a point scatterplot.
- `plot`: Creates a simple line chart.
- `bar`: Creates a bar chart.
- `hist`: creates a histogram.
- `boxplot` e `violinplot`: create figures that allow us to understand the distribution of the data.

In order to identify the figures, the following names will be used: 

- `fig`:reference to the general framework of the figure.
- `ax`ou `axs`: reference to subfigures created within a general framework.

## 1. Scatter Plot

```python
# Creating data
x = np.random.randn(1, 80)
y = np.random.randn(1, 80)

# Main figure
fig, ax = plt.subplots(figsize=(8, 8))

# Scatter plot
ax.scatter(x, y);
```

## 2. Line Graph

Para adicionar legendas e títulos, usaremos as funções a seguir:

```python
# Creating data
x = np.arange(1, 25, 1) # start stop step
y = x ** 2
z = x ** 3

# Main figure
fig, ax = plt.subplots(figsize=(6, 6))

# Line plot
ax.plot(x, y)
ax.plot(x, z)
```

To add subtitles and titles, we will use the following functions:

```python
# Creating data
x = np.arange(1, 25, 1)  # start stop step
y = x ** 2
z = x ** 3

# Main figure
fig, ax = plt.subplots(figsize=(8, 8))

# Line plot
ax.plot(x, y, label="$x^{2}$")
ax.plot(x, z, label="$x^{3}$")

# Adding legend
ax.legend()

# Adding title to the plot
ax.set_title('Exponential Graph', fontsize=14)

# Upper title - applied for the overall frame
fig.suptitle('Upper Title', fontsize=20)

# Adding x and y axis labels
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Adding grid lines to the plot
plt.grid(alpha=0.5)  # alpha is to make it slightly transparent
```

## 5. Boxplot

In some situations, there is a need to make modifications and stylizations to our graphics.

### 7.1Changing the point size 

To modify the size of a point on a scatterplot, we use the parameter `size`, which can also be abbreviated to `s`. Personally, I prefer to use the full function name, but this can change according to your preference.

```python
# Input
x = np.random.randn(1000)
y = 3*x + np.random.randn(1000)
z = np.random.randint(5, 100, (1000))

# Figura principal
fig, ax = plt.subplots(figsize=(8,8))

# scatterplot
im = ax.scatter(x,
                y,
                s = z
                );
```

```python
# Input
x = np.random.randn(1000)
y = 3*x + np.random.randn(1000)
z = np.random.randint(5, 100, (1000))

# Main figure
fig, ax = plt.subplots(figsize=(8,8))

# scatterplot
  #addittionally , we would change things like border and transparency
im = ax.scatter(x,
                y,
                s = z,
                alpha = 0.75,
                edgecolor = 'black'
                );
```

```python
# Main figure
fig, ax = plt.subplots(figsize=(8,8))

# Chart
im = ax.scatter(x, y, s=z, edgecolor="black", alpha=0.8)

#Defining the legend
handles, labels = im.legend_elements(prop="sizes")

# Writing the caption
ax.legend(title = 'Group Size',
          handles = handles, 
          labels = labels);
```

### 7.2 Color Bar

```python
# Input
x = np.random.randn(1000)
y = 3*x + np.random.randn(1000)
z = np.random.randn(1000)
size = np.random.randint(5, 100, (1000))

# Color map
cmap = plt.cm.inferno


# Main Figure
fig, ax = plt.subplots(figsize=(8,8))

# Chart
im = ax.scatter(x,
                y,
                c=z,
                cmap=cmap,
                s=50,
                edgecolor="black");

#Defining the legend
handles, labels = im.legend_elements(prop="colors")

# Writing the caption
ax.legend(title = 'Group Size',
          handles = handles, 
          labels = labels);
```

```python
# Main Figure
fig, ax = plt.subplots(figsize=(8,8))

# Chart
im = ax.scatter(x, y, c=z, cmap=cmap, s=50, edgecolor="black")

# Color Bar
cbar = fig.colorbar(im, ax=ax)
cbar.ax.set_title("Cu(%)", fontsize=18, y=1.01)
```
